```python
train_src_Y = train_y.to_numpy().reshape(-1,)
train_src_X = train_x.to_numpy()
test_X = test_x.to_numpy()

print(train_src_Y.shape)
print(train_src_X.shape)
print(test_X.shape)
```

```python
import sklearn.model_selection
import sklearn.preprocessing

train_X, val_X, train_Y, val_Y = sklearn.model_selection.train_test_split(
    train_src_X, train_src_Y, test_size = 0.25, random_state = 42,
    stratify = train_src_Y
)

print(train_X.shape)
print(val_X.shape)
print(train_Y.shape)
print(val_Y.shape)
train_X
```

```python
ss = sklearn.preprocessing.StandardScaler().fit(train_X)

train_X = ss.transform(train_X)
val_X = ss.transform(val_X)
test_X = ss.transform(test_X)
val_X
```

```python
import sklearn.linear_model
import sklearn.metrics


model = sklearn.linear_model.LogisticRegression().fit(train_X, train_Y)
pred = model.predict(val_X)
print(sklearn.metrics.f1_score(val_Y, pred, average = 'weighted'))
pred = model.predict(test_X)
pandas.DataFrame(pred, columns = [ 'price_range' ]).to_csv("result.csv", index = False)
```
