```python
wild = train[train['type'] == 'wild']
print(len(wild[wild['Pass'] == 1]) / len(wild['Pass']))

value1 = len(train.loc[(train['type'] == 'wild') & (train['Pass'] == 1)]) / len(train[train['type'] == 'wild'])
print(value1)
```

```python
domestic = train[train['type'] == 'domestic']
print(len(domestic[domestic['Pass'] == 1]) / len(domestic))

value2 = len(train.loc[(train['type'] == 'domestic') & (train['Pass'] == 1)]) / len(train[train['type'] == 'domestic'])
print(value2)
```

```python
abs(value1 - value2)
```
